중간값은 영어 < 수학 < 과학 으로 커진다.

과학은 중간값이 높다. 학생들이 과학시험에서 높은 점수를 받았다는 의미. 시험이 쉬웠다.

수학은 학생들끼리 점수 차이가 많이 난다. 영어랑 과학은 학생들끼리 비슷한 점수를 받았따.

영어는 중간값이 살짝 내려와있다. 25% ~ 50% 짧고, 50% ~ 75% 길다. 즉, 짧은 구간이 데이터가 더 빽빽하게 들어가있다. 넓은 부분은 데이터가 널찍하게 퍼져있다.

민지네 가족 구성원들의 키를 가지고 분산을 계산해 보세요. 민지의 아버지는 177cm, 민지의 오빠는 183cm, 민지의 어머니는 160cm, 민지는 164cm 입니다.

```python
father_height = 177
brother_height = 183
mother_height = 160
minzi_height = 164

avg = (father_height + brother_height + mother_height + minzi_height) / 4
avg
```

```python
father_height_var = (father_height - avg) ** 2
brother_height_var = (brother_height - avg) ** 2
mother_height_var = (mother_height - avg) ** 2
minzi_height_var = (minzi_height -  avg) ** 2

var = (father_height_var + brother_height_var + mother_height_var + minzi_height_var) / 4
var
```

```python
df = pd.read_csv('bike.csv')
df
```

```python
# bar 그래프
sns.set_theme(rc={'figure.figsize': (12, 6)}, style='white')  # 그래프 테마 설정
sns.barplot(data=df, x='month', y='total', errorbar=None)  #  자동으로 월별의 평균을 구해서 그래프로 그려줌
plt.show()
```

```python
sns.set_theme(rc={'figure.figsize': (12, 6)}, style='white')  # 그래프 테마 설정
sns.barplot(data=df, x='month', y='total', errorbar=None, hue='workingday')  # 총 대여 건수
plt.show()
```

```python
sns.set_theme(rc={'figure.figsize': (12, 6)}, style='white')  # 그래프 테마 설정
sns.barplot(data=df, x='month', y='registered', errorbar=None, hue='workingday')  # 정기권
plt.show()
```

정기권 있는 사람들은 워킹데이에 많이 탐. 즉, 출근이나 퇴근할 때 많이 이용한다고 예측 가능.

```python
sns.set_theme(rc={'figure.figsize': (12, 6)}, style='white')  # 그래프 테마 설정
sns.barplot(data=df, x='month', y='casual', errorbar=None, hue='workingday')  # 정기권 X
plt.show()
```

정기권 없이 자전거 타는 사람들은 주로 휴일에

```python
# line 그래프
sns.set_theme(rc={'figure.figsize': (12, 6)}, style='white')  # 그래프 테마 설정
sns.lineplot(data=df, x='month', y='casual', errorbar=None, hue='workingday')  # 정기권 X
plt.show()
```
