```python
def group_age(x):
    if x >= 10 and x < 20:
        return '10s'
    elif x >= 20 and x < 30:
        return '20s'
    elif x >= 30 and x < 40:
        return '30s'
    elif x >= 40 and x < 50:
        return '40s'
    elif x >= 50 and x < 60:
        return '50s'
    else:
        return '60s'
```

```python
patient_df['age_group_apply'] = patient_df['age'].apply(group_age)
patient_df
```

### 조인 연산

1. 공통으로 들어 있는 값을 기준으로 데이터를 합치는 것
2. key 값

**조인 연산 종류**
1. Inner Join: 공통 데이터 기준 합침
2. Left Outer Join: 왼쪽 데이터 기준 합침(오른쪽에 없는 데이터는 결측값이 됨)
3. Right Outer JOin: 오른쪽 데이터 기준 합침
4. Full Outer Join: 모든 데이터를 다 합침

```python
employee_df = pd.read_csv('data/employee.csv')
survey_df = pd.read_csv('data/survey.csv')
```

```python
employee_df
```

```python
survey_df
```

```python
# merge 함수의 기본은 Inner Join
pd.merge(employee_df, survey_df, on='id')
```

```python
pd.merge(employee_df, survey_df, on='id', how='left')
```
